# Map Visualization Incident Reports San Francisco

## Penjelasan Fitur
- IncidntNum: Nomor Insiden.
- Category: Kategori kejahatan atau kejadian.
- Descript: Deskripsi kejahatan atau kejadian.
- DayOfWeek: Hari dalam seminggu saat kejadian terjadi.
- Date: Tanggal terjadinya kejadian.
- Time: Waktu terjadinya peristiwa tersebut.
- PdDistrict: Departemen kepolisian distrik.
- Resolution: Penyelesaian tindak pidana ditinjau dari ditangkap atau tidaknya pelaku.
- Address: Alamat terdekat dengan tempat kejadian.
- X: Nilai garis bujur dari lokasi kejahatan.
- Y: Nilai garis lintang lokasi kejahatan.
- Location: Tupel nilai garis lintang dan garis bujur.
- PdId: ID departemen kepolisian.

## Menyiapkan Pustaka

```python
import pandas as pd
import numpy as np
import folium
```

```python
url = 'https://github.com/wildanmjjhd29/Visualization-Incident-Reports/releases/download/v1.0/dataset.csv'
incidents_df = pd.read_csv(url)
incidents_df.head()
```

```python
incidents_df.shape
```

```python
# Memabatasi dataset untuk mempercepat komputasi
incidents_df = incidents_df.iloc[0:100, :]
```

```python
incidents_df.shape
```

## Visualisasi Peta dengan Folium

```python
# tentukan peta dunia
world_map = folium.Map()
world_map
```

```python
world_map = folium.Map(location=[56.130, -106.35],zoom_start=3)
world_map
```

```python
# Nilai garis lintang dan garis bujur San Francisco
garis_lintang = 37.77
garis_bujur = -122.42

# membuat peta dan menampilkannya
sanfran_map = folium.Map(location=[garis_lintang,garis_bujur],zoom_start=12)

# menampilkan peta San Francisco
sanfran_map
```

```python
# membuat instance grup fitur untuk insiden dalam
incident = folium.FeatureGroup()

# ulangi 100 kejahatan dan tambahkan masing-masing ke grup fitur insiden
for lat,lng, in zip(incidents_df.Y, incidents_df.X):
  incident.add_child(
      folium.CircleMarker(
          [lat,lng],
          radius = 5,# tentukan seberapa besar penanda lingkaran yang Anda inginkan
          color = 'yellow',
          fill_color = 'blue',
          fill_opacity = 0.6
      )
)

# Tambahkan teks pop up ke setiap penanda di peta
lintang = list(incidents_df.Y)
bujur = list(incidents_df.X)
label = list(incidents_df.Category)

for lat,lng,label in zip(lintang,bujur,label):
  folium.Marker([lat,lng],popup=label).add_to(sanfran_map)

sanfran_map.add_child(incident)
sanfran_map
```

```python
# membuat peta dan menampilkannya
sanfran_map = folium.Map(location=[garis_lintang,garis_bujur],zoom_start=12)

# ulangi 100 kejahatan dan tambahkan masing-masing ke grup fitur insiden
for lat,lng, in zip(incidents_df.Y, incidents_df.X):
  incident.add_child(
      folium.CircleMarker(
          [lat,lng],
          radius = 5,# tentukan seberapa besar penanda lingkaran yang Anda inginkan
          color = 'yellow',
          popup=label,
          fill = True,
          fill_color = 'blue',
          fill_opacity = 0.6
      )
).add_to(sanfran_map)

sanfran_map
```

```python
from folium import plugins
```

## Menampilkan Peta Dengan Cluster

```python
sanfran_map = folium.Map(location=[garis_lintang,garis_bujur],zoom_start=12)

# Membuat instance object cluster untuk insiden
incident = plugins.MarkerCluster().add_to(sanfran_map)

for lat,lng,label in zip(incidents_df.Y,incidents_df.X, incidents_df.Category):
  folium.Marker(
      location=[lat,lng],
      icon =None,
      popup = label
  ).add_to(incident)

sanfran_map
```
